```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'year':[1900,1910,1920,1930,1940],
    'price':[250,375,425,500,690]
}
```

```python
data
```

```python
df = pd.DataFrame(data)
df.info()
```

```python
df.head()
```

```python
plt.scatter(df['year'],df['price'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
input_data = df['year'].values.reshape(-1,1)
output_data = df['price'].values.reshape(-1,1)
```

```python
model.fit(input_data,output_data)
```

```python
model.predict([[1990]])
```

```python
plt.scatter(df['year'],df['price'])
plt.plot(input_data,model.predict(input_data),color='red')
plt.show()
```
