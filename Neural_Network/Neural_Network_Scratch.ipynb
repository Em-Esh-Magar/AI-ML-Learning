```python
import numpy as np
import tensorflow as tf
```

```python
x = np.array([5.0])
w1 = np.array([2.9])
b1 = np.array([7.0])

w2 = np.array([5.4])
b2 = np.array([2.4])

# activation function
def relu(x):
  return np.maximum(0,x)
```

```python
# forward
a = np.dot(x,w1) + b1
z = relu(a)

y = np.dot(z,w2) + b2
y = relu(y)
print(y)
```

```python
# Loss Calculation
l = np.array([100.0])
loss = (l-y) ** 2
loss
```

```python
# backward
lr = 0.01

dl_dy = -2 * (l-y)
dl_dw = dl_dy * x
dl_db = dl_dy

new_weight = w1 - (lr * dl_dw)
new_bias = b1 - (lr * dl_db)
```

```python
print(new_weight)
print(new_bias)
```

```python
# epochs
for i in range(10):
  # forward
  a = np.dot(x,w1) + b1
  y = relu(a)

  loss = (l - y)**2
  # print(f"Loss : {loss}")

  # Backward
  dl_dy = -2 * (l-y)
  dl_dw = dl_dy * x
  dl_db = dl_dy

  w1 = w1 - (lr * dl_dw)
  w2 = b1 - (lr * dl_db)

  print(f"Epoch {i}, Loss:{loss}")
```
